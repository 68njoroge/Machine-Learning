```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Official column names for Boston Housing dataset
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS',
    'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]
df = pd.read_csv("House_price.csv",
                 sep='\s+',
                 names=column_names,
                 header=None
)
print("\n this is the shape:", df.shape)
print("\n the first three rows")
print(df.head(3))
print("\n colunm + missing values ")
print(df.info())
print("\n summary of numerical columns")
print(df.describe())
df.isnull().sum()
```

```python
#Handling missing values
#step 1 separate categorical and numerical
num_cols = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS',
    'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

#step 2 impute
num_imputer = SimpleImputer(strategy='mean')

#step 3 Apply them
df[num_cols] = num_imputer.fit_transform(df[num_cols])

#check missing values
df.isnull().sum()
```

```python
#StandardScaler
x = df.drop('MEDV', axis=1)
y = df['MEDV'].copy()
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
print('\n shape:',y.shape)
print('\n first three rows')
print(x.head(3))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2, random_state=42)
print("X_train shape:", x_train.shape)
print("X_test shape :", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

```python
#LinearRegression
#step 1 model creation
model = LinearRegression()
#step 2 train the model
model.fit(x_train,y_train)
#step 3 predict
y_pred = model.predict(x_test)
#evaluate using r2_score and mse
r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
print("R2 score:", r2)
print("MSE:", mse)
print("\n=== Model Performance ===")
print(f"R-squared (R²): {r2:.4f}")           # How well the model explains the variance (closer to 1 = better)
print(f"Mean Squared Error (MSE): {mse:.2f}")
#final coefficient
coefficients = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': model.coef_
})

print("\n=== Model Coefficients (how much each feature affects price) ===")
print(coefficients.sort_values(by='Coefficient', ascending=False).round(4))
```

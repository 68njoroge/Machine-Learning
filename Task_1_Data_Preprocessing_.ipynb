```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("1) iris.csv")
print("\n this is the shape:", df.shape)
print("\n the first three rows")
print(df.head(3))
print("\n colunm + missing values ")
print(df.info())
print("\n summary of numerical columns")
print(df.describe())
```

```python
#Handling missing values
#step 1 separate categorical and numerical
num_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
cat_cols = ['species']
#step 2 impute
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')
#step 3 Apply them
df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
#check missing values
df.isnull().sum()
```

```python
#Encoding categorical column
preprocessor = ColumnTransformer(transformers=[('cat', OneHotEncoder(handle_unknown= 'ignore', sparse_output=False),
['species']
)], remainder='passthrough')
#Apply
encoded = preprocessor.fit_transform(df)
print(preprocessor.get_feature_names_out())
```

```python
#StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(encoded)
print('\n first three rows of scaled data')
print(x_scaled[:3])
```

```python
#split data into training and testing set
y = df['species'].copy()
x_train,x_test,y_train,y_test = train_test_split(
    x_scaled, y, test_size=0.2, random_state=42
)
print("X_train shape:", x_train.shape)
print("X_test shape :", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```
